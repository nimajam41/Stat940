```python
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import os
```

```python
from dataset import load_data, save_data
from utils.utils import set_seed
```

```python
parent_dir = '/content/drive/MyDrive/stat940project'
data_path = os.path.join(parent_dir, 'data/ModelNet40')
```

```python
_, y = load_data(data_path, mode='test')
```

```python
def generate_target(label, NUM_CLASSES = 40):
    target = np.random.randint(NUM_CLASSES)
    while target == label:
        target = np.random.randint(NUM_CLASSES)
    return target
```

```python
generate_target = np.vectorize(generate_target)
targets = generate_target(y)
```

```python
assert (targets == y).any() == False
```

```python
save_data(targets, "test_targets", data_path)
```
